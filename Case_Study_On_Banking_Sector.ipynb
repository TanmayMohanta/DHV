```python
import pandas as pd

df = pd.DataFrame({
    'Age Group': ['18-25', '26-35', '36-45', '46-55', '55+'] * 4,
    'Total Loans': [100, 200, 300, 400, 500] * 4,
    'Number of Bad Loans': [5, 10, 15, 20, 25] * 4,
    'Number of Good Loans': [95, 190, 285, 380, 475] * 4
})

df['Percentage of Bad Loans'] = df['Number of Bad Loans'] / df['Total Loans'] * 100
df['Percentage of Good Loans'] = df['Number of Good Loans'] / df['Total Loans'] * 100

print(df)
```

```python
import matplotlib.pyplot as plt

max_loan_age_group = df.loc[df['Total Loans'].idxmax(), 'Age Group']

plt.figure(figsize=(8, 6))
plt.bar(df['Age Group'], df['Total Loans'], color='skyblue')
plt.title('Total Loans by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Loans')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.bar(max_loan_age_group, df.loc[df['Age Group'] == max_loan_age_group, 'Total Loans'], color='red')

plt.show()
```

```python
import pandas as pd
import numpy as np

np.random.seed(0)
age_group = np.random.choice(['18-25', '26-35', '36-45', '46-55', '56-65'], size=20)
total_loans = np.random.randint(10, 100, size=20)
bad_loans = np.random.randint(0, total_loans, size=20)
good_loans = total_loans - bad_loans
percentage_bad_loans = (bad_loans / total_loans) * 100
percentage_good_loans = (good_loans / total_loans) * 100

banking_data = pd.DataFrame({
    'Age Group': age_group,
    'Total Loans': total_loans,
    'Bad Loans': bad_loans,
    'Good Loans': good_loans,
    '% Bad Loans': percentage_bad_loans,
    '% Good Loans': percentage_good_loans
})

print(banking_data)
```

```python
import matplotlib.pyplot as plt

loan_by_age = banking_data.groupby('Age Group')['Total Loans'].sum()

plt.figure(figsize=(10, 6))
loan_by_age.plot(kind='bar', color='skyblue')
plt.title('Total Loans by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Loans')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.bar(max_loan_age_group, df.loc[df['Age Group'] == max_loan_age_group, 'Total Loans'], color='red')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.bar(banking_data.index, banking_data['Bad Loans'], label='Bad Loans', color='magenta')

plt.bar(banking_data.index, banking_data['Good Loans'], bottom=banking_data['Bad Loans'], label='Good Loans', color='cyan')

plt.xlabel('Index')
plt.ylabel('Number of Loans')
plt.title('Good Loans vs Bad Loans')
plt.legend()
plt.show()
```
