```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "date":pd.to_datetime(["2023-01-01","2023-02-01","2023-03-01","2023-04-01","2023-05-01","2023-06-01","2023-07-01","2023-08-01","2023-09-01","2023-10-01","2023-11-01"]),
"value":[10,15,22,28,35,42,48,55,60,68,75]
    }
df=pd.DataFrame(data)
df=df.set_index("date",inplace=False)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='v',linestyle='--')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series With Trend")
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
data = {
    "Date": pd.date_range(start='2022-01-01', periods=365),
    "Value": [20 + 10 * ((i % 30) // 7) for i in range(365)]
}
df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
result = seasonal_decompose(df["Value"], model='additive', period=30)
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(df.index, result.trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(df.index, result.seasonal, label='Seasonality')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(df.index, result.resid, label='Trend and Seasonality')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
