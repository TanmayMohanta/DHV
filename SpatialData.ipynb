```python
import folium

mymap = folium.Map(location=[13.17189450907666, 77.5361273121627],
                   width=950,
                   height=550,
                   zoom_start=12,
                   tiles='openstreetmap')

mymap
```

```python
marker = folium.Marker(
    location=[12.927097831594144, 77.64963225296839],
    popup='Shri Jagannath Temple',
    icon=folium.Icon(color='red')
)

mymap.add_child(marker)

mymap
```

```python
import folium

mymap = folium.Map(location=[13.013371307375166, 77.58810505900468], zoom_start=12)

image_url = 'https://images.unsplash.com/photo-1688878181284-532867d4263e?w=500&auto=format&fit=crop&q=60&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxzZWFyY2h8Mnx8YmVuZ2FsdXJ1JTIwcGFsYWNlfGVufDB8fDB8fHww'

icon = folium.CustomIcon(
    image_url,
    icon_size=(50, 50),
    icon_anchor=(25, 25),
)

marker = folium.Marker(
    location=[13.010878845912375, 77.587252342819],
    popup='Bengaluru Palace',
    icon=icon
)

mymap.add_child(marker)

mymap
```
